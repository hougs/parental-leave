```python
import pandas as pd
import seaborn as sb

%matplotlib inline
```

```python
leave = pd.read_csv("parental-leave.csv")
parental_leave = leave[leave["Parental Class"] == "Maternity"].sort("Base in Weeks")
ordered_comps = parental_leave.Company.values.tolist()
```

```python
plot = sb.barplot("Company", y="Base (Full Pay) in Weeks", hue="Parental Class", data=leave, x_order=ordered_comps)
plot.set_xticklabels(ordered_comps, rotation=30, ha='right')
plot
```

```python
def keyed_df_by_parental_class(pclass):
    df = parental_leave[parental_leave["Parental Class"] == pclass].set_index("Company")
    del df["Parental Class"]
    return df

p_leave = keyed_df_by_parental_class("Paternity")
m_leave = keyed_df_by_parental_class("Maternity")
```

```python
percent_df = pd.DataFrame(index=m_leave.index)
percent_df["frac_p_leave"] = p_leave["Base in Weeks"]/m_leave["Base in Weeks"]
percent_df["frac_m_leave"] = m_leave["Base in Weeks"]/p_leave["Base in Weeks"]
percent_df["maternity_leave"] = m_leave["Base in Weeks"]
percent_df.plot(x="Paternity:Maternity Ratio", y="Maternity Leave (weeks)", kind='scatter', xlim=(0.0, 1.14))
```
